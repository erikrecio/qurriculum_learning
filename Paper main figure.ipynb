```python
import numpy as np
import matplotlib.pyplot as plt

n_datapoints = 100
width = 0.6
plt.rcParams['font.size'] = 20
color = "darkred"

x = np.arange(n_datapoints)
score = np.random.randint(1, 100, n_datapoints)
score_sorted = np.sort(score)


plt.bar(x, score, width=width, color=color)
plt.tick_params(left = False, right = False , labelleft = False , 
                labelbottom = False, bottom = False)
plt.xlabel("Training samples")
plt.ylabel("Score")
plt.savefig("test.svg")
plt.show()

plt.bar(x, score_sorted, width=width, color=color)
plt.tick_params(left = False, right = False , labelleft = False , 
                labelbottom = False, bottom = False)
# plt.xlabel("Training samples")
plt.ylabel("Score")
plt.savefig("test_sorted.svg")
plt.show()
```
